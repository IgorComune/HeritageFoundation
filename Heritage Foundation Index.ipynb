# Property Rights and PPP

```python
rp_ppp = countries_region[['Region','Country Name','Property Rights','GDP per Capita (PPP)']]
rp_ppp
```

```python
rp_ppp_table = rp_ppp.groupby('Region').mean().sort_values('Property Rights', ascending=False)
rp_ppp_table.round(2)
```
